```python
class Peca:
    def __init__(self, id, pos_x, pos_y, icone):
        self.id = id
        self.pos_x = pos_x
        self.pos_y = pos_y
        self.coordenada = [pos_x, pos_y]  # Representa a coordenada atual da Peca
        self.icone = icone
        self.removida = False

    def __str__(self):
        return f"{id}"
```

```python
def inicializa_pecas(
        mapeamento_posicao_inicial: dict[list[int, int]], 
        mapeamento_icones: dict[list[str, str]]
    ) -> list[Peca]:
    
    lista_pecas_criadas = []

    for id in mapeamento_posicao_inicial.keys():

        indice_icone = [id[0] == x[0] for x in mapeamento_icone_pecas].index(True)
        icone = mapeamento_icone_pecas[list(mapeamento_icone_pecas.keys())[idx]]

        lista_pecas_criadas.append(
            Peca(id
                 , mapeamento_posicao_inicial[id][1]
                 , mapeamento_posicao_inicial[id][0]
                 , icone[0 if id[1] == "P" else 1]
                )
            )
    return lista_pecas_criadas
```

```python
l = inicializa_pecas(mapeamento_posicao, mapeamento_icone_pecas)

for i in l:
    print(f"{i.id} - {i.coordenada} - {i.icone}")
```
